# Data Cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("./raw_practo.csv")
df.head()
```

```python
df.shape
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
for i in df:
    print(i)
    print(df[i].nunique())
    print(df[i].unique())
```

```python
df['Year_of_experience']=df['Year_of_experience'].replace('[^0-9]',' ',regex=True).\
str.strip().str.split(expand=True)[0].astype(float)
```

```python
df['npv']=df['npv'].replace("[^0-9]",'',regex=True).astype(int)
```

```python
df['consultation_fee']=df['consultation_fee'].replace("[^0-9]",'',regex=True).astype(int)
```

```python
df['dp_score']=df['dp_score'].replace("[^0-9]",'',regex=True).astype(int)
```

```python
df.head()
```

```python
df[['Location', 'ABC']] = df['Location'].str.split(',', expand=True)
```

```python
df.drop('ABC',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Year_of_experience'].isna().sum()
```

```python
df=df.dropna()
```

```python
df['Year_of_experience']=df['Year_of_experience'].astype(int)
```

```python
df.dtypes
```

```python
df['npv(votes)']=df['npv']
```

```python
df.drop('npv',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df.shape
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df['Degree_']=df['Degree'].str.split(expand=True)[0]
```

```python
df['Degree_']=df['Degree_'].str.replace("[^a-zA-Z/]",'',regex=True)
```

```python
df['Degree_'].unique()
```

```python
df['Degree_'].replace('DO','Doctorate',inplace=True)
df['Degree_'].replace('Doctor','Doctorate',inplace=True)
df['Degree_'].replace('Post','PG',inplace=True)
```

```python
remove = ['FNB', 'Fellow', 'DVD', 'Member', 'MRCOGUK', 'MRCS', 'DNHE','Fellowship','Masters']
df = df[~df['Degree_'].isin(remove)]
```

```python
# df['Dgree-1'].value_counts().tail(8)
```

```python
df['Degree_'].unique()
```

```python
df.drop('Degree',axis=1,inplace=True)
```

```python
df.rename(columns={'Degree_': 'Degree'}, inplace=True)
```

```python
df.drop('Name',axis=1,inplace=True)
```

```python
df.reset_index()
```

```python
df.dtypes
df
```

```python
df.to_csv('clean_practo_data.csv',index=False)
```

## City Wise analysis of doctors

```python
x=df2.groupby('City').agg({'Speciality':'count'}).sort_values(by='Speciality',ascending=False).reset_index()
x
```

```python
cities = ['Bangalore', 'Delhi', 'Mumbai']
values = x['Speciality']

colors = ['#32CD32', '#9370DB', '#FFA500']

plt.figure(figsize=(6, 4))
bars = plt.bar(cities, values, color=colors,data=x)

#Data labels
for bar, value in zip(bars, values):
    plt.annotate(str(value), (bar.get_x() + bar.get_width() / 2, bar.get_height() + 5),
                 ha='center', va='bottom')

plt.xlabel('City')
plt.ylabel('Total Number of Doctors')
plt.title('City-wise Total Number of Doctors', y=1.02, fontsize=12,fontweight='bold')

plt.tight_layout()
plt.show()
```

## Top 5 Most Common Degrees

```python
x=df2.groupby('Degree').agg({'City':'count'}).sort_values(by='City',ascending=False).head(5)
x
```

```python
categories = ['MBBS', 'MDS', 'BDS', 'MD', 'BPTh/BPT']
values = [3362, 873, 712, 203, 169]

colors = ['#FF8C00', '#8A2BE2', '#008000', '#FF4500', '#6495ED']

plt.figure(figsize=(6, 5))
wedges, texts, autotexts = plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=140, colors=colors,data=x)

for i, (wedge, text, autotext) in enumerate(zip(wedges, texts, autotexts)):
    autotext.set_text(f'{values[i]}')
    autotext.set_color('white')

plt.title('Top 5 Degrees', y=1.02, fontsize=12,fontweight='bold')
plt.axis('equal')
plt.show()
```

## Count of Doctors in different Specialties

```python
x=df2.groupby('Speciality').agg({'City':'count'}).sort_values(by='City',ascending=False).reset_index()
x
```

```python
# Create a bar chart
fig, ax = plt.subplots(figsize=(10, 6))

plt.bar(x['Speciality'], x['City'], color='darkgrey')

plt.xlabel('Speciality')
plt.ylabel('Count')
plt.title('Count of Specialities (Descending Order)')
plt.xticks(rotation=30, ha='right')

for container in ax.containers:
    for bar in container:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, yval + 0, f'{yval:.0f}', ha='center', va='bottom', color='black', \
                 fontsize=10)

plt.tight_layout()
plt.show()
```

## Maximun and Minimum price

```python
x = pd.DataFrame(df2.groupby('Speciality')['consultation_fee'].agg(['max','min'])).sort_values(by='max',ascending=False)
x=x.reset_index()
x
```

```python
# Create a clustered column chart
fig, ax = plt.subplots(figsize=(10, 6))

bar_width = 0.35
index = range(len(x))

plt.bar(index, x['max'], bar_width, label='Max Consultation Fee', color='g', alpha=0.6)
plt.bar([i + bar_width for i in index], x['min'], bar_width, label='Min Consultation Fee', color='r', alpha=0.6)

plt.xlabel('Speciality')
plt.ylabel('Consultation Fee')
plt.title('Max and Min Consultation Fees by Speciality',y=1.02, fontsize=12,fontweight='bold')
plt.xticks([i + bar_width/2 for i in index], x['Speciality'], rotation=30, ha='right')
plt.legend()

for container in ax.containers:
    for bar in container:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, yval + 10, f'{yval:.0f}', ha='center', color='black', fontsize=8)

plt.tight_layout()
plt.show()
```

## Average years of experience

```python
x = pd.DataFrame(df2.groupby('Speciality')['Year_of_experience'].agg('mean')).sort_values(by='Year_of_experience',ascending=False).reset_index()
x
```

```python
# Create a column chart
fig, ax = plt.subplots(figsize=(10, 6))
plt.bar(x['Speciality'], x['Year_of_experience'], color='purple')

for container in ax.containers:
    for bar in container:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, yval + 0, f'{yval:.0f}', ha='center', va='bottom', color='black', \
                 fontsize=10)
plt.xlabel('Speciality')
plt.ylabel('Mean Year of Experience')
plt.title('Mean Year of Experience by Speciality')
plt.xticks(rotation=30, ha='right')

plt.tight_layout()
plt.show()
```
